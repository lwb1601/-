## numpy array 생성하기

ndarray 또는 배열(array)는 연산에 특화된 자료형이다.  
형태는 리스트와 비슷하고 인덱싱과 슬라이싱도 가능하다.  
차이점은 array는 하나의 데이터 타입만 원소로 가질 수 있으며  반복문을 사용하지 않고 쉽고 빠른 연산이 가능하다.


````
# 리스트를 활용하여 배열 생성

import numpy as np
a = np.array([1,2,3,4])
print(a)
````

````
# 인덱싱 슬라이싱

print(a[-1])
print(a[1:3])
````

````
# 하나의 배열의 모든 원소는 데이터 타입이 같아야한다.

a = np.array([1,2,3,'4'])
print(a)
````

### 4. 궁금한 지역의 인구구조와 가장 비슷한 인구 구조를 가진 지역 찾기

비슷한 인구구조는 어떻게 찾는가?  
방법1) 연령별 인구 수 차이의 합이 가장 적은 지역 찾기  
방법2) 연령별 인구수 비율의 차이의 합이 가장 적은 지역 찾기

알고리즘(방법1)

1.   전국의 모든 지역 중 한 곳을 선택한다.
2.   궁금한 지역 A의 0세 인구수에서 B의 0세 인구수를 뺀다.
3.   2.를 100세 이상 인구수에 해당하는 값까지 반복한 후 각각의 차이를 모두 더한다.
4. 전국의 모든 지역에 대해 반복하며 그 차이가 가장 작은 지역을 찾는다.




````

import csv
import numpy as np
import matplotlib.pyplot as plt



f = open('/content/age.csv', encoding='cp949') # 천단위 콤마가 제거된 데이터 csv데이터 사용
data = csv.reader(f)
next(data)

data=list(data)

name = input('인구구조가 알고 싶은 지역의 이름을 입력해 주세요: ')

home =[]
away =[]

min = 99999

result =[]
result_name = 0


for row in data:
  if name in row[0]:
    home = np.array(row[3:], dtype=int)
    name = row[0]

for row in data:
  away = np.array(row[3:], dtype=int)
  temp = np.sum(abs(home-away))
  if temp != 0 and temp < min:
    min = temp
    result = away
    result_name = row[0]

plt.rc('font', family = 'NanumBarunGothic')
plt.title(name+' 지역과 가장 비슷한 인구 구조를 가진 지역')
plt.plot(home, label=name)
plt.plot(result, label=result_name)
plt.legend()
plt.show()

````
